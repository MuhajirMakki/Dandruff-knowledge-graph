**cleaning**

```python
# remove extra spaces and standardize text
df['Subject'] = df['Subject'].str.strip().str.title()
df['Object'] = df['Object'].str.strip().str.title()
df['Relationship'] = df['Relationship'].str.strip().str.lower()
```

```python
df.head()
```

**Filter CORE KNOWLEDGE GRAPH EDGES only**

```python
core_relations = [
    "causes",
    "associated_with",
    "treats",
    "produces",
    "activates",
    "inhibits",
    "breaks_down",
    "damages",
    "protects"
]

df_core = df[df['Relationship'].isin(core_relations)]

df_core.head()
```

**Extract Disease-centered graph**

```python
diseases = ["Dandruff", "Seborrheic Dermatitis"]

df_disease = df_core[
    (df_core['Subject'].isin(diseases)) |
    (df_core['Object'].isin(diseases))
]

df_disease
```

**filtering Microbiome network**

```python
microbes = [
    "Malassezia", "Staphylococcus", "Streptococcus",
    "Cutibacterium", "Candida", "Aspergillus", "Filobasidium"
]

df_microbes = df_core[
    (df_core['Subject'].isin(microbes)) |
    (df_core['Object'].isin(microbes))
]

df_microbes
```

**filtering Drug network**

```python
drugs = [
    "Ketoconazole", "Zinc Pyrithione", "Selenium Sulfide",
    "Fluconazole", "Itraconazole", "Ciclopirox"
]

df_drugs = df_core[
    (df_core['Subject'].isin(drugs)) |
    (df_core['Object'].isin(drugs))
]

df_drugs
```

**Build “Knowledge Graph Edges**

```python
kg_edges = df_core[[
    "Subject",
    "Relationship",
    "Object",
    "Subject_Type",
    "Object_Type",
    "Source_ID"
]]

kg_edges.head()
```

**Basic insight**

```python
print("Total relations:", len(df))
print("Core relations:", len(df_core))
print("Disease-related:", len(df_disease))
print("Microbe-related:", len(df_microbes))
print("Drug-related:", len(df_drugs))
```

**Networkx work onward**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
## this create graph
G = nx.DiGraph()
```

**Add nodes + edges**

```python
for _, row in kg_edges.iterrows():
    subject = row["Subject"]
    obj = row["Object"]
    relation = row["Relationship"]
    
    # Add nodes with type info
    G.add_node(subject, type=row["Subject_Type"])
    G.add_node(obj, type=row["Object_Type"])
    
    # Add edge with full biomedical metadata
    G.add_edge(
        subject,
        obj,
        relation=relation,
        source=row["Source_ID"]
    )
print(list(G.nodes(data=True))[:5])
print(list(G.edges(data=True))[:5])
```

**graph summary**

```python
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

**Disease-centered graph**

```python
diseases = {"Dandruff", "Seborrheic Dermatitis"}

disease_nodes = set()

for u, v in G.edges():
    if u in diseases or v in diseases:
        disease_nodes.add(u)
        disease_nodes.add(v)

G_disease = G.subgraph(disease_nodes)
```

**Microbiome network**

```python
microbes = {
    "Malassezia", "Staphylococcus", "Streptococcus",
    "Cutibacterium", "Candida", "Aspergillus", "Filobasidium"
}

microbe_nodes = set()

for u, v in G.edges():
    if u in microbes or v in microbes:
        microbe_nodes.add(u)
        microbe_nodes.add(v)

G_microbe = G.subgraph(microbe_nodes)
```

**Drug network**

```python
drugs = {
    "Ketoconazole", "Zinc Pyrithione", "Selenium Sulfide",
    "Fluconazole", "Itraconazole", "Ciclopirox"
}

drug_nodes = set()

for u, v in G.edges():
    if u in drugs or v in drugs:
        drug_nodes.add(u)
        drug_nodes.add(v)

G_drug = G.subgraph(drug_nodes)
```

**VISUALIZATION**

**Biomedical Knowledge Graph**

```python
import matplotlib.pyplot as plt
import networkx as nx

plt.figure(figsize=(16, 12))

#Keep meaningful relations
important_relations = ["causes", "treats", "associated_with"]

G_best = nx.DiGraph()

for u, v, data in G.edges(data=True):
    if data.get("relation") in important_relations:
        
         
        if G.nodes[u].get('type') in ["Chemical", "Process"]:
            continue
        if G.nodes[v].get('type') in ["Chemical", "Process"]:
            continue

        G_best.add_edge(u, v, **data)
        G_best.nodes[u]['type'] = G.nodes[u].get('type')
        G_best.nodes[v]['type'] = G.nodes[v].get('type')

#Better spacing layout
pos = nx.spring_layout(G_best, k=2.8, iterations=200, seed=42)

 
color_map = []
for node, data in G_best.nodes(data=True):
    ntype = data.get('type', '')
    
    if ntype == "Disease":
        color_map.append("#ff6b6b")  # soft red
    elif ntype == "Microorganism":
        color_map.append("#51cf66")  # green
    elif ntype == "Drug":
        color_map.append("#339af0")  # blue
    else:
        color_map.append("#ced4da")  # gray

# here we Draw nodes
nx.draw(
    G_best,
    pos,
    node_color=color_map,
    with_labels=False,
    node_size=3200,
    edge_color="gray",
    arrows=True,
    width=1.5
)

#  Clean labels (bigger & bold)
nx.draw_networkx_labels(
    G_best,
    pos,
    font_size=11,
    font_weight='bold'
)

#  No edge labels 


plt.title("Biomedical Knowledge Graph (Microbes -> Disease -> Treatment)", fontsize=16)

plt.axis('off')
plt.show()
```

## saving KG to owl,rdf extensions files

```python
import re
import networkx as nx
from rdflib import Graph, Namespace, URIRef, Literal, RDF, RDFS, OWL, XSD
```

```python
# Choose the graph to export
if "G_full" not in globals():
    raise NameError("G_full not found. Run the cell that creates G_full first.")

graph_to_export = G_full
```

```python
# 2) Namespaces
EX = Namespace("http://example.org/biomedical_kg/")
RES = Namespace("http://example.org/biomedical_kg/resource/")

rdf_graph = Graph()
rdf_graph.bind("ex", EX)
rdf_graph.bind("res", RES)
rdf_graph.bind("rdf", RDF)
rdf_graph.bind("rdfs", RDFS)
rdf_graph.bind("owl", OWL)
rdf_graph.bind("xsd", XSD)
```

**Helper functions**

```python
def safe_iri(text):
    """
    Convert text to a safe URI fragment.
    Keeps letters, numbers, underscore, hyphen, and dot.
    """
    text = str(text).strip()
    text = re.sub(r"\s+", "_", text)
    text = re.sub(r"[^A-Za-z0-9_\-\.]", "", text)
    if not text:
        text = "item"
    return text

def to_literal(value):
    """
    Convert Python value to rdflib Literal with basic datatype handling.
    """
    if isinstance(value, bool):
        return Literal(value, datatype=XSD.boolean)
    if isinstance(value, int):
        return Literal(value, datatype=XSD.integer)
    if isinstance(value, float):
        return Literal(value, datatype=XSD.decimal)
    return Literal(str(value))

```

**Declare ontology**

```python
ontology_uri = EX.BiomedicalKnowledgeGraph
rdf_graph.add((ontology_uri, RDF.type, OWL.Ontology))
rdf_graph.add((ontology_uri, RDFS.label, Literal("Biomedical Knowledge Graph")))
```

**Collect node types and relations**

```python
node_types = set()
relation_names = set()

for node, data in graph_to_export.nodes(data=True):
    node_types.add(data.get("type", "Entity"))

for u, v, data in graph_to_export.edges(data=True):
    relation_names.add(data.get("relation", "related_to"))
```

**Create OWL classes for node types**

```python
for ntype in sorted(node_types):
    class_uri = EX[safe_iri(ntype)]
    rdf_graph.add((class_uri, RDF.type, OWL.Class))
    rdf_graph.add((class_uri, RDFS.label, Literal(ntype)))


# Create object properties for relations

for rel in sorted(relation_names):
    prop_uri = EX[safe_iri(rel)]
    rdf_graph.add((prop_uri, RDF.type, OWL.ObjectProperty))
    rdf_graph.add((prop_uri, RDFS.label, Literal(rel)))

# Generic relation property too
rdf_graph.add((EX.related_to, RDF.type, OWL.ObjectProperty))
rdf_graph.add((EX.related_to, RDFS.label, Literal("related_to")))
```

**Add nodes as individuals**

```python
for node, data in graph_to_export.nodes(data=True):
    node_uri = RES[safe_iri(node)]
    node_type = data.get("type", "Entity")
    class_uri = EX[safe_iri(node_type)]

    rdf_graph.add((node_uri, RDF.type, OWL.NamedIndividual))
    rdf_graph.add((node_uri, RDF.type, class_uri))
    rdf_graph.add((node_uri, RDFS.label, Literal(node)))

    # preserve node attributes
    for attr_key, attr_value in data.items():
        if attr_value is None:
            continue
        attr_predicate = EX[safe_iri(attr_key)]
        rdf_graph.add((node_uri, attr_predicate, to_literal(attr_value)))
```

**Add edges as RDF triples**

```python
for u, v, data in graph_to_export.edges(data=True):
    subj_uri = RES[safe_iri(u)]
    obj_uri = RES[safe_iri(v)]

    rel = data.get("relation", "related_to")
    prop_uri = EX[safe_iri(rel)]

    # Main relation triple
    rdf_graph.add((subj_uri, prop_uri, obj_uri))

    # Also add a generic relation triple
    rdf_graph.add((subj_uri, EX.related_to, obj_uri))

    # Preserve edge metadata
    for attr_key, attr_value in data.items():
        if attr_value is None:
            continue
        meta_predicate = EX[safe_iri(f"{rel}_{attr_key}")]
        rdf_graph.add((subj_uri, meta_predicate, to_literal(attr_value)))
```

**Save files**

```python
rdf_graph.serialize(destination="biomedical_knowledge_graph.owl", format="xml")      # OWL/RDF XML
rdf_graph.serialize(destination="biomedical_knowledge_graph.rdf", format="xml")      # RDF/XML
rdf_graph.serialize(destination="biomedical_knowledge_graph.ttl", format="turtle")   # Turtle
rdf_graph.serialize(destination="biomedical_knowledge_graph.nt", format="nt")        # N-Triples

print("Saved files successfully:")
print(" - biomedical_knowledge_graph.owl")
print(" - biomedical_knowledge_graph.rdf")
print(" - biomedical_knowledge_graph.ttl")
print(" - biomedical_knowledge_graph.nt")
print("Nodes exported:", graph_to_export.number_of_nodes())
print("Edges exported:", graph_to_export.number_of_edges())
```
